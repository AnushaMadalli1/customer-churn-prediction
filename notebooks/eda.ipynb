# EDA Notebook for Customer Churn Prediction

Load data, explore distributions, correlations, churn rates.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load raw data
df = pd.read_csv('../data/raw/churn.csv')  # Update path
df.head()
```

```python
# Churn rate
churn_rate = df['Churn'].value_counts(normalize=True)
churn_rate.plot(kind='bar')
plt.title('Churn Rate')
plt.show()
```
